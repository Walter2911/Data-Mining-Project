```python
import matplotlib.pyplot as plt

monthly_data = border_data.groupby(border_data['Date'].dt.to_period("M")).agg({'Value': 'sum'})
monthly_data.index = monthly_data.index.to_timestamp()

#plotting crossing over time
plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data['Value'])
plt.title('Total Border Crossings Over Time')
plt.xlabel('Date')
plt.ylabel('Total Crossings')
plt.grid(True)
plt.show()
```

### Now let's move onto the next step in cleaning which if finding and resolving the outliers.

*) To identify the outliers we will first calculate the Interquartile Range(IQR)

*) Any data point is considered an outlier if:
    Value < Q1 - 1.5 * IQR
    Value > Q3 + 1.5 * IQR
    

```python
# Calculate Q1, Q3, and IQR
Q1 = border_data['Value'].quantile(0.25)
Q3 = border_data['Value'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
outliers = border_data[(border_data['Value'] < lower_bound) | (border_data['Value'] > upper_bound)]

# Counting the number of outliers
num_outliers = outliers.shape[0]

lower_bound, upper_bound, num_outliers, outliers.head()
```

By the results we can see that the lower bound is -4440 and the upper bound is 7400 and we also have detected a 72,665 rows of 394901 rows where Value is outside this range.

*) Next we create a box plot showing overall distribution and highlights extreme values as "points outside the box"

*) Scatter plot over time which helps visualize how outliers behave over the years.

```python
plt.figure(figsize=(10, 6))
plt.boxplot(border_data['Value'], vert=False, patch_artist=True, showmeans=True)
plt.title('Box Plot of Value Column (Border Crossings)')
plt.xlabel('Value')
plt.show()
```

*) By looking at the boxplot we can see thta the data is right skewed which means there are few extremly high values that dominate the dataset.

*) We can also say that the majority of the values data points are clustered tightly near the lower end of the range(close to 0).

*) And also a massive number of outliers extend far to the right, with some exceeding 1*10^6. This shows that a few extreme crossing volumes are driving the variability in the dataset.

*key-point for presentation:*

-> The box plot shows that the majority of the border crossing volumes are concentrated below 7400, with a meadian value near 0. However, there is a significant number of outliers extending to over 1 million crossings, indicating potential anomolies.

-> Instead of removing these velues, we will analyze them further using the clustering methods to identify meaninful patterns.

```python
plt.figure(figsize=(12, 6))
plt.scatter(border_data['Date'], border_data['Value'], alpha=0.5, s=10)
plt.title('Scatter Plot of Border Crossings Over Time')
plt.xlabel('Date')
plt.ylabel('Value (Border Crossings)')
plt.grid(True)
plt.show()
```

*) Border crossing volumes are significantlly reduced during this perid, likely due to pandemic(COVID-19) and associated travel restrictions. This aligns with the sharp drop we noticed in the time series plot.

There are extreme values in the box plot. They may correspond to special events or anomolies worth further investigation.

*) There's a gradual recovery after 2022, but it hasent yet reached the high levels observed in earlier years.

*Key points for presentation*:

-> he scatter plot shows that border crossing volumes increased steadily from 1996 to 2004, with a peak in activity during that period. A significant drop occurred between 20202022, likely due to the COVID-19 pandemic.

-> Several extreme values exceed 1 million crossings, particularly before 2004. These represent unusual spikes in activity, likely tied to special events or trade surges. These outliers will be analyzed further using clustering to determine if they are meaningful anomalies.

```python
high_value_outliers = border_data[border_data['Value'] > 7400]

outlier_summary = high_value_outliers[['Date', 'Port Name', 'State', 'Measure', 'Value']].sort_values(by='Value', ascending=False)

outlier_summary.head()
```

*) The highest border crossing volumes occured at the El Paso port in texas, primarly for personal vehicles passengers between 2000 and 2001.

*) The most extreme outlier was in march 2001 with 4+ million crossings.


```python
# Aggregate data by 'Date' and 'Measure' to observe temporal patterns
temporal_agg = border_data.groupby(['Date', 'Measure'])['Value'].sum().reset_index()

temporal_agg.head()
```

```python
specific_period = temporal_agg[(temporal_agg['Date'] >= '2000-01-01') & (temporal_agg['Date'] <= '2001-12-31')]
specific_period
```

```python
plt.figure(figsize=(12, 6))
for measure in specific_period['Measure'].unique():
    subset = specific_period[specific_period['Measure'] == measure]
    plt.plot(subset['Date'], subset['Value'], label=measure)
plt.title('Temporal Trends for 20002001')
plt.xlabel('Date')
plt.ylabel('Value (Border Crossings)')
plt.legend()
plt.grid(True)
plt.show()
```

*key points for presentation*:

-> The temporal trends for 2000-2001 show that Personal Vehicle Passengers dominate border activity, contributing significantly to the highest crossing volumes."

-> While Pedestrians exhibit steady traffic during this period, measures like Personal Vehicle Passengers fluctuate, potentially reflecting seasonal or event-driven changes.

-> The extreme values in Personal Vehicle Passengers align with the identified outliers, reinforcing their significance for anomaly detection
