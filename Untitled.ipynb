First we start by standardizing the text in categorial columns.

Starting with border and measure as they ar the key columns that affect analysis because they describe the category of the crossing like Us-Mexico vs Us-Canada" border or the type of crossing.

```python
# Display unique values in Border and Measure columns
unique_borders = border_data['Border'].unique()
unique_measures = border_data['Measure'].unique()

print("Unique values in Border column:", unique_borders)
print("Unique values in Measure column:", unique_measures)
```

```python
# Convertig only the first letter in the categorial columns into uppercase and rest of the value into lowercase.

# For port name
border_data['Port Name'] = border_data['Port Name'].str.title()

# For Measure
border_data['Measure'] = border_data['Measure'].str.title()

# For State
border_data['State'] = border_data['State'].str.title()

# For Border
border_data['Border'] = border_data['Border'].str.title()


#print(border_data['Port Name'])
#print(border_data['Measure'])
#print(border_data['State'])
#print(border_data['Border'])
```

If we see clearly we can observe that the date data is categorial which means it can be in either string format or text-numerical format. Now let's convert that to proper datatime format.

```python
# Coverting date column into proper datetime format
border_data['Date'] = pd.to_datetime(border_data['Date'], format='%b %Y')  # Example format

print(border_data['Date'].dtype)
```

```python
# print(border_data.dtypes)
```

```python
import matplotlib.pyplot as plt

# Aggregate total crossings by month
monthly_data = border_data.groupby(border_data['Date'].dt.to_period("M")).agg({'Value': 'sum'})
monthly_data.index = monthly_data.index.to_timestamp()  # Convert PeriodIndex to Timestamp for plotting

# Plot total crossings over time
plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data['Value'])
plt.title('Total Border Crossings Over Time')
plt.xlabel('Date')
plt.ylabel('Total Crossings')
plt.grid(True)
plt.show()
```
